```python
menu_items = ["Popcorn", "Candy Bar", "Soda", "Nachos", "Pretzel"]
item_prices = [3.50, 2.75, 2.00, 4.00, 3.25]
inventory = {
    "Popcorn": 10,
    "Candy Bar": 15,
    "Soda": 20,
    "Nachos": 8,
    "Pretzel": 12
}

print("Snack Bar Menu:")
for i in range(len(menu_items)):
    print(f"{i + 1}. {menu_items[i]} - ${item_prices[i]:.2f}")
order = []
print("escolha o numero do item que você deseja comprar (ou digite 'done' para finalizar):")
while True:
    try:
        choice_input = input("Enter your choice: ")
        if choice_input.lower() == 'done':
            break
        choice = int(choice_input)
        item_index = choice - 1

        if 0 <= item_index < len(menu_items):
            item_name = menu_items[item_index]
            if item_name in inventory and inventory[item_name] > 0:
                order.append(item_index)
                inventory[item_name] -= 1
                print(f"{item_name} added to your order. Remaining stock: {inventory[item_name]}")
            elif item_name in inventory and inventory[item_name] == 0:
                print(f"Sorry, {item_name} is out of stock.")
            else:
                print(f"Error: {item_name} not found in inventory.")
        else:
            print("numero de item invalido, tente novamente.")
    except ValueError:
        print("Invalid input. Please enter a number or type 'done'.")
print("\nOrder Summary:")
for item_index in order:
    print(f"- {menu_items[item_index]} - ${item_prices[item_index]:.2f}")

total_cost = 0
for item_index in order:
    total_cost += item_prices[item_index]

print(f"\nTotal: ${total_cost:.2f}")
```
