## Criar a função de soma

### Subtask:
Definir uma função que receba três números como parâmetros e retorne a soma deles.


**Reasoning**:
The subtask is to define a function that sums three numbers. This can be done in a single Python function definition.



```python
def calcular_soma(num1, num2, num3):
  """Calcula a soma de três números."""
  return num1 + num2 + num3
```

## Criar a função de média

### Subtask:
Definir uma função que receba a função de soma como parâmetro (ou os três números diretamente e chame a função de soma internamente) e retorne a média.


**Reasoning**:
Define the `calcular_media` function that takes three numbers, calls `calcular_soma` internally, and returns the average.



```python
def calcular_media(num1, num2, num3):
  soma = calcular_soma(num1, num2, num3)
  media = soma / 3
  return media
```

## Tratar exceções na entrada do usuário

### Subtask:
Implementar tratamento de exceções (por exemplo, `try-except`) para lidar com entradas não numéricas.


**Reasoning**:
Implement exception handling for non-numeric input within the menu function as per the instructions.



```python
def calcular_soma(num1, num2, num3):
  return num1 + num2 + num3

def calcular_media(num1, num2, num3):
  soma = calcular_soma(num1, num2, num3)
  media = soma / 3
  return media

def menu():
  while True:
    print("\nEscolha uma opção:")
    print("1 - Calcular Soma e Média")
    print("2 - Sair")

    escolha = input("Digite o número da opção desejada: ")

    if escolha == '1':
      try:
        # Prompt the user for three numbers
        num1_str = input("Informe o primeiro número: ")
        num2_str = input("Informe o segundo número: ")
        num3_str = input("Informe o terceiro número: ")

        # Convert the inputs to numbers
        num1 = float(num1_str)
        num2 = float(num2_str)
        num3 = float(num3_str)

        # Calculate and display the sum and average (This part will be added later)
        soma = calcular_soma(num1, num2, num3)
        media = calcular_media(num1, num2, num3)
        print(f"\nA soma é: {soma}")
        print(f"A média é: {media}")

      except ValueError:
        print("\nEntrada inválida. Por favor, digite apenas números.")

    elif escolha == '2':
      print("Saindo do programa.")
      break
    else:
      print("Opção inválida. Por favor, escolha 1 ou 2.")

# Call the menu function to start the program
menu()
```
